Opening csv file

```python
import urllib.request
url = 'https://raw.githubusercontent.com/covid19-eu-zh/covid19-eu-data/master/dataset/covid-19-at.csv'
filename = 'covid-19-at.csv'
urllib.request.urlretrieve(url, filename)
```

Reading csv file

```python
import csv

with open('covid-19-at.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        print(', '.join(row))
```

Número de mortes em Tirol

```python
with open('covid-19-at.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    list_cases = []
    list_recovered = []
    list_deaths = []
    list_tests = []
    list_hospitalized = []
    list_intensive_care = []
    for row in reader:
        line = row
        if line[1] == "Tirol":
            list_cases.append(int(line[2]))
            list_recovered.append(int(line[3]))
            list_deaths.append(int(line[4]))
            list_tests.append(int(line[5]))
            list_hospitalized.append(int(line[6]))
            list_intensive_care.append(int(line[7]))   
    cases = sum(list_cases)
    recovered = sum(list_recovered)
    deaths = sum(list_deaths)
    tests = sum(list_tests)
    hospitalized = sum(list_hospitalized)
    intensive_care = sum(list_intensive_care)
    with open('covid-19-austria.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["name", "cases", "recoverd", "deaths", "tests", "hospitalized", "intensive care"])
        writer.writerow(["Tirol", str(cases), str(recovered), str(deaths), str(tests), str(hospitalized), str(intensive_care)])
    
```
